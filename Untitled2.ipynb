```python
include = df[np.logical_and(df["favorite_count"]>=1, df["retweet_count"]>=1)]
```

```python
include.head()
```

```python
f = []
for row in include['favorite_count']:
    if row < 10.62:
        f.append('low')
    elif row > 10.62:
        f.append('high')
```

```python
r = []
for row in include['retweet_count']:
    if row < 18.09:
        r.append('low')
    elif row > 18.09:
        r.append('high')
```

```python
include['f'] = f
```
